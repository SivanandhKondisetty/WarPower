```python
import plotly.express as px
close=raw['Close']
normalized = (close / close.iloc[0]) * 100
normalized.columns=['Crude oil','Gold','S&P 500 index']
fig = px.line(
    normalized,
    title="WarPower — US-Iran War 2026: Resource Impact"
 )

normalized.columns=['oil','gold','stocks']
normalized.head()
fig.add_vline(x='2026-02-28', line_width=2, line_dash="dash", line_color="red",annotation_text="US-Iran War Begins", annotation_position="top left")
fig.add_vline(x='2026-04-08', line_width=2, line_dash="dash", line_color="Green",annotation_text="Ceasefire Declared", annotation_position="top right")
fig.show(renderer="browser")
```

```python
import plotly.express as px
import pandas as pd
import yfinance
raw2 = yfinance.download(['NG=F','ZW=F','LMT','JPY=X'],start="2025-01-01", end="2026-06-10")
close2=raw2['Close']
normalized2 = (close2 / close2.iloc[0]) * 100
normalized2.columns = ['USD/JPY', 'Lockheed Martin', 'Natural Gas', 'Wheat']
normalized2.loc['2026-01-28', 'Natural Gas'] = None
normalized2['Natural Gas'] = normalized2['Natural Gas'].ffill()
fig2=px.line(
    normalized2,
    title="WarPower — US-Iran War 2026: Resource Impact (Energy, Food, Defense, Currency)"
)
fig2.add_vline(x='2026-02-28', line_width=2, line_dash="dash", line_color="red",annotation_text="US-Iran War Begins", annotation_position="top left")
fig2.add_vline(x='2026-04-08', line_width=2, line_dash="dash", line_color="Green",annotation_text="Ceasefire Declared", annotation_position="top right")
fig2.show()
```

```python
import pandas as pd
peak_date = normalized2['Natural Gas'].idxmax()
print(normalized2['Natural Gas'].loc[peak_date - pd.Timedelta(days=5) : peak_date + pd.Timedelta(days=5)])
```

```python
all_data= pd.concat([normalized,normalized2],axis=1,sort=False)
before= all_data.loc['2026-02-27']
after= all_data.loc['2026-03-30']
pct_change = ((after - before) / before) * 100
print(pct_change)
```
